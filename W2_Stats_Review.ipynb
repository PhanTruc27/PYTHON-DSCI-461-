## Find the average value of one-bedroom apartments in Manhattan, Brooklyn, Queens

```python
# Read in housing data
brooklyn = pd.read_csv('2brooklyn.csv')
manhattan = pd.read_csv('2manhattan.csv')
queens = pd.read_csv('2queens.csv')
```

```python
# Filter value for one-bedroom
brooklyn['bedrooms'].value_counts() # 340
manhattan['bedrooms'].value_counts() #  1476
queens['bedrooms'].value_counts() #232
```

```python
brooklyn_1br_price = brooklyn.loc[brooklyn['bedrooms'] == 1, 'rent']
manhattan_1br_price = manhattan.loc[manhattan['bedrooms'] == 1, 'rent']
queens_1br_price = queens.loc[queens['bedrooms'] == 1, 'rent']
```

```python

print("Brooklyn mean price:", brooklyn_mean_1br)
print("Manhattan mean price:", manhattan_mean_1br)
print("Queens mean price:", queens_mean_1br)
```

## A trip to London - Let's check the weather

```python

import seaborn as sns
```

```python
london_data = pd.read_csv('2london_weather.csv')
```

```python
london_data.head()
```

```python
# Read date type data of date
london_data.info()
```

```python
london_data['date'] = pd.to_datetime(
    london_data['date'],
    format='%Y%m%d'
)

london_data['date']
```

```python
london_data.info()
```

```python
#create cols of year, month, day to save the values
london_data['year'] = london_data['date'].dt.year
london_data['month'] = london_data['date'].dt.month
london_data['day'] = london_data['date'].dt.day
```

```python
london_data['year'].unique()
```

```python
# Data for May
may_data = london_data[london_data['month'] == 5]
may_data
```

```python
london_data.columns
```

```python
# Variance of temperature in May
may_data_var =may_data.loc[may_data['year']>=2015]['mean_temp'].agg(['var','std'])
may_data_var
```

```python
# Mean of temperature in May
may_data_mean = may_data.loc[may_data['year']>=2015]['mean_temp'].mean()
round(may_data_mean,2)
```

```python
# Data for June
june_data = london_data[london_data['date'].dt.month == 6]

# Variance of temperature in June
june_data_var = june_data.loc[june_data['year']>=2015]['mean_temp'].agg(['var','std'])
june_data_var
```

```python
# Mean of temperature in June
june_data_mean = june_data.loc[june_data['year']>=2015]['mean_temp'].mean()
june_data_mean
```

```python
# Create df to save Month, Variance, Mean

DATE = pd.DataFrame({
    'Month': ['May','Jun'],
   'Variance': [june_data_var, may_data_var],
    'Mean':[may_data_mean, june_data_mean
    ]
})

DATE
```

```python
print("Average temperature in May: " + str(may_data_mean))
print("Average temperature in Jun: " + str(june_data_mean))
# Temperature in May is cooler than in Jun
print("Standard deviation of temperature in London in May: " + str(may_data_var))
print("Standard deviation of temperature in London in Jun: " + str(june_data_var))
# And the temperature in May is much unstable then in Jun (the standard deviation is larger)
```

# 3. Quartiles, Quantiles, IRQ

Quantiles are values that split a dataset into groups of equal.

**Common Quantiles**  
One of the most common quantiles is the 2-quantile. This value splits the data into two groups of equal size.  
This is also known as the median!

![Common Quantiles](Images/median.svg)

**Quartiles and Interquartile**  
The 4-quantiles, or the quartiles, split the data into four groups of equal size.


![interquartile](Images/interquartile.svg)

The interquartile range is the different between the third (Q3) and the first quartile (Q1). The Q1 is the value that separates the first 25% of the data from the remaining 75%. The Q3 is the opposite - it separates the first 75% of the data from the remaining 25%.
IQR, like the range, helps describe the spread of the center of the data. But, unlike the range, the IQR is robust. A statistics is robust when outliers have little impact on it.



**Percentiles**  
The values that splits the data into 100 groups, are commonly used to compare new data points to the dataset.


![Quartiles](Images/quartiles.svg)

```python
from scipy.stats import iqr
```

```python
dataset = [4, 10, 38, 85, 193]
```

```python
#Calculate q1, q3, iqr_range
q1=np.quantile(dataset,0.25)
```

```python
q3=np.quantile(dataset,0.75)
```

```python
iqr_range =q3-q1
```

```python

interquartile_range = iqr(dataset)
interquartile_range
```

```python
# IQR is robust
dataset_one = [6, 9, 10, 45, 190, 200]
dataset_two = [6, 9, 10, 45, 190, 20000000]
```

```python
print(iqr(dataset_one), iqr(dataset_two))
```

```python
print(max(dataset_one) - min(dataset_one))
print(max(dataset_two) - min(dataset_two))
```
